### Задание на повторение материала

```python
a = np.linspace(12, 24, num=12, endpoint=False)
a
```

```python
def QuantityColons(q, x):
    if len(q) % x == 0:
        y = len(q)/x
    else:
        y1 = len(q)//x
        w = len(q) - (x * y1)
        q = q[:-int(w)]
        y = len(q) / x
    z = q.reshape(x, int(y))
    s = f'Размер нового массива {z.shape}'
    return z, s
```

```python
# 1-массив
QuantityColons(a, 2)
```

```python
# 2-й массив
QuantityColons(a, 3)
```

```python
# 3-й массив
QuantityColons(a, 4)
```

```python
# 4-й массив
QuantityColons(a, 5)
```

```python
# 5-й массив
QuantityColons(a, 6)
```

```python
# 6-й массив (для проверки)
QuantityColons(a, 7)
```
